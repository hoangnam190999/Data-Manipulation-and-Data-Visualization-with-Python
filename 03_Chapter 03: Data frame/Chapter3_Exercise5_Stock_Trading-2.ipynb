```python
# Question 1c: Read the companies.csv file => put data into companies
companies = pd.read_csv('stock_trading_data/companies.csv')
# View companies data
companies.head()
```

```python
# Question 2: Indicate whether there is Null data in stocks1 or not?
stocks1.isnull().sum()
```

```python
stocks1[stocks1.isnull().any(axis=1)]
```

```python
stocks1[['symbol','high']]
```

```python
stocks1.groupby('symbol')['high'].transform(max)
```

```python
# If yes, replace with the following rule:
# If 'high' column is Null then replace with max value on 'high' column of that stock code
# If 'low' column is Null then replace with min value on 'low' column of that stock code
stocks1.high = stocks1.high.fillna(stocks1.groupby('symbol')['high'].transform('max'))
stocks1.low = stocks1.low.fillna(stocks1.groupby('symbol')['low'].transform('min'))
```

```python
# Question 3: Create a stocks dataframe by merging stocks1 and stocks2 by row
stocks = pd.concat([stocks1,stocks2],join='inner',ignore_index=True)
# View the last 15 rows of stocks
stocks.tail(15)
```

```python
# Question 4: Create dataframe stocks_companies by merging stocks and companies
stocks_companies = pd.merge(stocks,companies,left_on='symbol',right_on='name',how='inner')
# View the first 5 rows of data of stocks_companies
stocks_companies.head()
```

```python
# Question 7: Create a parsed_time column in stocks_companies by converting time to DateTime format
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
# Specify the data type of the parsed_time column
stocks_companies['parsed_time'].dtype
# Display the first 5 rows of data in stocks_companies
stocks_companies.head()
```

```python
# Question 8: Add result column, if price 'close' > 'open' then result column has value 'up', otherwise 'down'
stocks_companies['result'] = np.where(stocks_companies.close > stocks_companies.open,'up','down')
stocks_companies
```

```python
# Question 9: Draw a line chart showing the trading volume trend of AMZN and comment
stocks_companies.loc[stocks_companies.name=='AMZN',['date','volume']].plot(x='date')
```

```python
# Question 10: Draw a line chart showing the trend of FB's closing price and comment
stocks_companies.loc[stocks_companies.name=='FB',['date','close']].plot(x='date')
```
