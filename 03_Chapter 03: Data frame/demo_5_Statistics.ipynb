### Melt

```python
pivot = pivot.reset_index()
pivot
```

```python
unpivoted_df = pd.melt(pivot, id_vars=['DEPARTMENT_NAME'], var_name='HIRE_YEAR_tmp', value_name='mean_SALARY')

# Delete NaN
unpivoted_df = unpivoted_df.dropna()

# Display unpivoted DataFrame
unpivoted_df
```
