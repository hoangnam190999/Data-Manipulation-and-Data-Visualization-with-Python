

- map:  
map được sử dụng cho Series (cột) và hoạt động với từng phần tử.  
Nó ánh xạ một giá trị trong Series với một giá trị khác.

- apply:  
apply hoạt động trên Series hoặc trên toàn cột hoặc dòng trong DataFrame.  
Nó cho phép bạn áp dụng một hàm tự định nghĩa cho mỗi phần tử hoặc dòng/cột.

- applymap:  
applymap hoạt động trên toàn DataFrame và áp dụng một hàm cho từng ô (cell) trong DataFrame.  
Nó cho phép bạn thực hiện biến đổi dữ liệu trên toàn DataFrame.

```python
data = {
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': ['apple', 'banana', 'cherry', 'date']
}

df = pd.DataFrame(data)
df['A'] = df['A'].map({1: 'One', 2: 'Two', 3: 'Three', 4: 'Four'})
df
```

```python
def ktsnt(n):
    if n > 1:
        for i in range(2,n):
            if n %i==0:
                return False
        else: 
            return True
    else:
        return False
```

```python
df['D'] = df['B'].map(lambda x: ktsnt(x))
df
```

```python
df['D1'] = df['B'].apply(ktsnt)
df
```

```python
df_upper = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
df_upper
```
