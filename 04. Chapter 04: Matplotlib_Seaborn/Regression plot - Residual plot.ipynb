```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import random
x = [random.randint(0, 20) for i in range(13)]
print(x)
```

```python
y = [3*i+1 for i in x]
print(y)
```

```python
plt.subplot(1,2,1)
sns.regplot(x = x, y = y)
plt.title('regplot')

plt.subplot(1,2,2)
sns.residplot(x = x, y = y)
plt.title('residplot')
plt.ylim(-0.5, 0.5)
plt.suptitle('Đa thức bậc 1')
plt.show()
```

```python
x = [random.randint(-20, 20) for i in range(20)]
print(x)
```

```python
y = [i**2 - 3*i+1 for i in x]
print(y)
```

```python
df = pd.DataFrame({'x': x, 'y': y})
df.head()
```

```python
plt.subplot(1,3,1)
sns.regplot(x = x, y = y, order=2)
plt.title('regplot', fontsize = 8)

plt.subplot(1,3,2)
sns.residplot(x = x, y = y)
plt.title('residplot vs order = 1 (default)', fontsize = 8)

plt.subplot(1,3,3)
sns.residplot(x = x, y = y, order=2)
plt.title('residplot vs order = 2', fontsize = 8)
plt.ylim(-0.5, 0.5)

plt.suptitle('Đa thức bậc 2')
plt.tight_layout()
plt.show()
```
